```python
#Buatlah sebuah fungsi untuk membuat matrix berdasarkan input dari user

def create_matrix():
    rows = int(input("Masukkan jumlah baris: "))
    cols = int(input("Masukkan jumlah kolom: "))

    matrix = []

    for i in range(rows):
        row = []
        for j in range(cols):
            element = float(input(f"Masukkan elemen untuk baris {i+1}, kolom {j+1}: "))
            row.append(element)
        matrix.append(row)

    return matrix

# Contoh penggunaan fungsi
matrix = create_matrix()
print("Matrix yang dibuat:")
for row in matrix:
    print(row)
```

```python
#Buatlah sebuah fungsi untuk menambah dua matrix berdasarkan input user

def tambah_matrix():
    # Membaca ukuran matriks dari pengguna
    baris = int(input("Masukkan jumlah baris: "))
    kolom = int(input("Masukkan jumlah kolom: "))

    # Membaca elemen-elemen matriks pertama dari pengguna
    print("Masukkan elemen-elemen matriks pertama:")
    matrix1 = []
    for i in range(baris):
        row = []
        for j in range(kolom):
            elemen = int(input(f"Masukkan elemen [{i+1}][{j+1}]: "))
            row.append(elemen)
        matrix1.append(row)

    # Membaca elemen-elemen matriks kedua dari pengguna
    print("Masukkan elemen-elemen matriks kedua:")
    matrix2 = []
    for i in range(baris):
        row = []
        for j in range(kolom):
            elemen = int(input(f"Masukkan elemen [{i+1}][{j+1}]: "))
            row.append(elemen)
        matrix2.append(row)

    # Menampilkan matriks pertama
    print("Matriks Pertama:")
    for row in matrix1:
        print(row)

    # Menampilkan matriks kedua
    print("Matriks Kedua:")
    for row in matrix2:
        print(row)

    # Menambahkan kedua matriks
    hasil = []
    for i in range(baris):
        row = []
        for j in range(kolom):
            elemen = matrix1[i][j] + matrix2[i][j]
            row.append(elemen)
        hasil.append(row)

    # Menampilkan hasil penjumlahan matriks
    print("Hasil Penjumlahan:")
    for row in hasil:
        print(row)
        
# Memanggil fungsi untuk menjalankan program
tambah_matrix()
```

```python
#Buatlah fungsi untuk mengkalikan dua buah matrix berdasarkan input user

def multiply_matrices():
    # Membaca ukuran matriks dari input pengguna
    rows1 = int(input("Masukkan jumlah baris matriks pertama: "))
    cols1 = int(input("Masukkan jumlah kolom matriks pertama: "))
    rows2 = int(input("Masukkan jumlah baris matriks kedua: "))
    cols2 = int(input("Masukkan jumlah kolom matriks kedua: "))

    # Validasi ukuran matriks
    if cols1 != rows2:
        print("Error: Jumlah kolom matriks pertama harus sama dengan jumlah baris matriks kedua.")
        return

    # Membaca elemen-elemen matriks pertama dari input pengguna
    print("Masukkan elemen-elemen matriks pertama:")
    matrix1 = []
    for i in range(rows1):
        row = []
        for j in range(cols1):
            element = float(input(f"Masukkan elemen matriks pertama [{i}, {j}]: "))
            row.append(element)
        matrix1.append(row)

    # Membaca elemen-elemen matriks kedua dari input pengguna
    print("Masukkan elemen-elemen matriks kedua:")
    matrix2 = []
    for i in range(rows2):
        row = []
        for j in range(cols2):
            element = float(input(f"Masukkan elemen matriks kedua [{i}, {j}]: "))
            row.append(element)
        matrix2.append(row)

    # Mengalikan matriks pertama dengan matriks kedua
    result_matrix = []
    for i in range(rows1):
        row = []
        for j in range(cols2):
            element = 0
            for k in range(cols1):
                element += matrix1[i][k] * matrix2[k][j]
            row.append(element)
        result_matrix.append(row)

    # Menampilkan matriks pertama
    print("\nMatriks Pertama:")
    for row in matrix1:
        print(row)

    # Menampilkan matriks kedua
    print("\nMatriks Kedua:")
    for row in matrix2:
        print(row)

    # Menampilkan hasil perkalian matriks
    print("\nHasil Perkalian Matriks:")
    for row in result_matrix:
        print(row)
                
# Memanggil fungsi untuk menjalankan program
multiply_matrices()
```
